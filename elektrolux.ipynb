```python
file = open('response_elektrolux.json', 'r')
data = json.load(file)
file.close()
```

```python
index_main = []
index_sub = []

for item in data['products']:
    for spec in item['specifications']:
        if spec['name'] == 'CookingFunctions':
            for item in spec['value'].split('|'):
                val = 'Tryb_' + item
                if val not in index_sub:
                    index_main.append('Performance')
                    index_sub.append(val)

for item in data['products']:
    for spec in item['specifications']:
        if spec['name'] != 'CookingFunctions':
            if spec['translatedName'] not in index_sub:
                index_main.append(spec['groupName'])
                index_sub.append(spec['translatedName'])

for item in data['products']:
    for spec in item['extendedSpecifications']:
        if spec['translatedName'] not in index_sub:
                index_main.append(spec['groupName'])
                index_sub.append(spec['translatedName'])

for item in data['products']:
    for f in item['features']:
        val = f['bullet'] if 'bullet' in f else f['title']
        if val not in index_sub and val.startswith('Funkcje') == False:
            index_main.append('Features')
            index_sub.append(val)

index_main.append('Other')
index_sub.append('Cena')
index_main.append('Other')
index_sub.append('Nazwa')
arrays = [
    np.array(index_main),
    np.array(index_sub)
]
pairs = list(zip(index_main, index_sub))
sorted_pairs = sorted(pairs, key=lambda x: (x[0], x[1]))
sorted_arrays = [np.array([x[0] for x in sorted_pairs]), np.array([x[1] for x in sorted_pairs])]
sorted_arrays
```

```python
model_products = []
for item in data['products']:
    model_products.append(item['modelId'])
model_products = list(set(model_products))
model_products
```

```python
df = pd.DataFrame('---', index=sorted_arrays, columns=model_products, dtype='str')
# df.loc[('Performance', 'Tryb_Au Gratin'), 'KODDP77H'] = 'TAK'
df
```

```python
for item in data['products']:
    for spec in item['specifications']:
        if spec['name'] == 'CookingFunctions':
            for s in spec['value'].split('|'):
                val = 'Tryb_' + s
                df.loc[('Performance', val), item['modelId']] = "TAK"
        else:
            df.loc[(spec['groupName'], spec['translatedName']), item['modelId']] = spec['value'].replace('|','\n')
df.head()
```

```python
for item in data['products']:
    for spec in item['extendedSpecifications']:
        df.loc[(spec['groupName'], spec['translatedName']), item['modelId']] = spec['value']
df.head()
```

```python
for item in data['products']:
    if 'price' in item:
        df.loc[('Other', 'Cena'), item['modelId']] = item['price']['formattedValue']
    df.loc[('Other', 'Nazwa'), item['modelId']] = item['name']
```

```python
for item in data['products']:
    for f in item['features']:
        val = f['bullet'] if 'bullet' in f else f['title']
        if not val.startswith('Funkcje'):
            df.loc[('Features', val), item['modelId']] = 'TAK'
```
